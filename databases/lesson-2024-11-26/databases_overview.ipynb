## Introduction to Databases

## What is a database? 
Before we get into what a database is all about, let's take a look at what `data` is.

### What is data
> Data can be defined as a representation of facts, concepts or instructions in a formalized manner.

For example

```
23
31
Alice 
34
John 
Bob 27
Corinne 
```

### Early days of data

What do we usually do with data?
- Managing data
- Preserve data

## Database
> A database is a systematic collection of data for effective processing.

We have two types of databases
- `Flat file database (Single File) / non-relational database`
    - `plain text files` like `.txt` files, `.csv` files.
    ```csv
    name,age
    Alice,34 
    John,31
    Bob,27
    Corinne,23
    ```
- `Relational` / `Structured database`
It contains tables of data with rows and columns.

|name|age|
|----|---|
|Alice|34|
|John|31|
|Bob|27|
|Corinne|23|







## Why database?
Databases help us to effectively

## DBMS - Database Management System
> A DBMS is a software for creating and managing databases.


### Example of DBMS
- `DB2` -> IBM
- `Oracle` -> Oracle
- `SQL Server` -> Microsoft
- `MySQL`
- `PostgreSQL`.

**NB**: We will focus more on `Relational DBMS`, the DMS for relational databases.

We have 3 more common types of DBMS 
- `Hierarchical DBMS`
- `network DBMS`
- `object-oriented DBMS`.

### Benefits of DBMS
- `Data security`
- `Data integrity`
- `crash recovery`
- `concurrency control`
- `Access control`

## Relational Data Modeling

### Relational Schema
It is a like a blueprint, describing the structure of the table. Using with the the table name, attributes, constraints, domains.

`Student(student_id INT PRIMARY KEY, name VARCHAR(100), age INT)`


### Key Constraint
Two things about Primary Keys
- They are unique
- They are not null.

### Domain Constraints
This implements the attribute domain that states that every attribute should be bound to have a value that lies inside a specific range of values.
- Age -> `18 - 25`
- Domain constraint also involves `data type`.
    - int, float, string
